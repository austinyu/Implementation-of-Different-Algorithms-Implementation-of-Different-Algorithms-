```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("experiment.csv")
df
```

```python
size = df.iloc[:,0]
average = df.iloc[:, 4]
```

```python
m, b = np.polyfit(size, average, 1, full = True)[0]
res = np.polyfit(size, average, 1, full = True)[4]
x = np.arange(size.min(), size.max(), .1)
plt.figure(figsize = (20, 10))
plt.plot(x, m * x + b, c = 'r')
plt.scatter(size, average, s = 5)
plt.suptitle('The Number of Swaps Operation vs Array Size')
plt.xlabel('Array Size')
plt.ylabel('The Number of Swaps Operation')
print(f'The sum of squared differences for linear approximation is {res}.')
```
