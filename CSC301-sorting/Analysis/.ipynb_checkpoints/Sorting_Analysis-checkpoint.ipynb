```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
bucketSort = pd.read_csv('competition-bucketSort.csv', index_col = 0, header = None).iloc[:, :-1].T
insertionSort = pd.read_csv('competition-insertionSort.csv', index_col = 0, header = None).iloc[:, :-1].T
radixSort = pd.read_csv('competition-radixSort.csv', index_col = 0, header = None).iloc[:, :-1].T
recMergeSort = pd.read_csv('competition-recMergeSort.csv', index_col = 0, header = None).iloc[:, :-1].T
timSort = pd.read_csv('competition-timSort.csv', index_col = 0, header = None).iloc[:, :-1].T
```

```python
# Insertion sort
insertionSort_stack = pd.DataFrame(insertionSort.stack())
size = [i[1] for i in np.array(insertionSort_stack.index)]
time = [i[0] for i in insertionSort_stack.values]
insertionSort_stack = pd.DataFrame({'size' : size,
             'time' : time})
insertionSort_stack['sort'] = 'insertionSort'

# Recsive MergeSort
recMergeSort_stack = pd.DataFrame(recMergeSort.stack())
size = [i[1] for i in np.array(recMergeSort_stack.index)]
time = [i[0] for i in recMergeSort_stack.values]
recMergeSort_stack = pd.DataFrame({'size' : size,
             'time' : time})
recMergeSort_stack['sort'] = 'recMergeSort'

# Tim Sort
timSort_stack = pd.DataFrame(timSort.stack())
size = [i[1] for i in np.array(timSort_stack.index)]
time = [i[0] for i in timSort_stack.values]
timSort_stack = pd.DataFrame({'size' : size,
             'time' : time})
timSort_stack['sort'] = 'timSort'

# Bucket Sort
bucketSort_stack = pd.DataFrame(bucketSort.stack())
size = [i[1] for i in np.array(bucketSort_stack.index)]
time = [i[0] for i in bucketSort_stack.values]
bucketSort_stack = pd.DataFrame({'size' : size,
             'time' : time})
bucketSort_stack['sort'] = 'bucketSort'

# Radix Sort
radixSort_stack = pd.DataFrame(radixSort.stack())
size = [i[1] for i in np.array(radixSort_stack.index)]
time = [i[0] for i in radixSort_stack.values]
radixSort_stack = pd.DataFrame({'size' : size,
             'time' : time})
radixSort_stack['sort'] = 'radixSort'
```

```python
merged = pd.merge(recMergeSort_stack, insertionSort_stack, how="outer")
plt.figure(figsize = (20, 10))
sns.violinplot(x= "size", y= "time", hue = 'sort',
                    data = merged, scale = 'width')
plt.title('The Comparison between Time Complexities of Recurssive Merge Sort and Insertion Sort', fontsize = 20)
plt.xlabel('Array Size', fontsize = 15)
plt.ylabel('Time', fontsize = 15)
```

```python
merged = pd.merge(recMergeSort_stack, radixSort_stack, how="outer")
plt.figure(figsize = (20, 10))
sns.violinplot(x= "size", y= "time", hue = 'sort',
                    data = merged, scale = 'width')
plt.title('The Comparison between Time Complexities of Recurssive Merge Sort and Radix Sort', fontsize = 20)
plt.xlabel('Array Size', fontsize = 15)
plt.ylabel('Time', fontsize = 15)
```

```python
merged = pd.merge(recMergeSort_stack, timSort_stack, how="outer")
plt.figure(figsize = (18, 9))
sns.violinplot(x= "size", y= "time", hue = 'sort',
                    data = merged, scale = 'width')
plt.title('The Comparison between Time Complexities of Recurssive Merge Sort and Tim Sort', fontsize = 20)
plt.xlabel('Array Size', fontsize = 15)
plt.ylabel('Time', fontsize = 15)
```

```python
merged = pd.merge(radixSort_stack, bucketSort_stack, how="outer")
plt.figure(figsize = (20, 10))
sns.violinplot(x= "size", y= "time", hue = 'sort',
                    data = merged, scale = 'width')
plt.title('The Comparison between Time Complexities of Radix Sort and Bucket Sort', fontsize = 20)
plt.xlabel('Array Size', fontsize = 15)
plt.ylabel('Time', fontsize = 15)
```
